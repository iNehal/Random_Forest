```python
from sklearn.preprocessing import LabelEncoder
```

```python
data['time']=encoder.fit_transform(data['time'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
x_train.info()
```

```python
cat_col=['sex','smoker','day']
num_col=['total_bill','tip','size']
```

```python
num_pipeline=Pipeline(
    
    steps=[
    
        ('imputer',SimpleImputer(strategy='median')),
        ('scaler',StandardScaler())
    
    ]

)

cat_pipeline=Pipeline(

    steps=[
    
        ('imputer',SimpleImputer(strategy='most_frequent')),
        ('encoder',OneHotEncoder())
    
    
    ]

)
```

```python
preprocessor=ColumnTransformer([

    ('num_pipeline',num_pipeline,num_col),
    ('cat_pipeline',cat_pipeline,cat_col)

])
```

```python
x_train=preprocessor.fit_transform(x_train)
x_test=preprocessor.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
models={

    'random_forest':RandomForestClassifier(oob_score=True),
    'logistic_regression':LogisticRegression(),
    'decision_tree':DecisionTreeClassifier()

}
```

```python
from sklearn.metrics import accuracy_score
def evaluate_model(x_train,y_train,x_test,y_test,models):
    report={}
    for i in range(len(models)):
        model=list(models.values())[i]
        model.fit(x_train,y_train)
        print (model.score(x_train,y_train))
        #print("OOB score:", model.oob_score_)
        
        y_pred=model.predict(x_test)
        
        accuracy=round(accuracy_score(y_test,y_pred)*100,2) #round means no. ko round off krna
        
        report[list(models.keys())[i]]=accuracy
    return report
```

```python
list(models.values())[1]
```

```python
evaluate_model(x_train,y_train,x_test,y_test,models) #pehla result training accuracy h and 2nd result testing accuracy h
```
